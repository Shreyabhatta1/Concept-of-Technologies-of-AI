```python
import pandas as pd
# Load the dataset
file_path='/content/drive/MyDrive/Data Set/Copy of medical_students_dataset.csv'
df=pd.read_csv(file_path)
print(df.info())
#Check for missing values.
missing_values=df.isnull().sum()
print("Missing values in each columns:\n",missing_values)
#Handling missing values.
df['Age'].fillna(df['Age'].mean())
df['Height'].fillna(df['Height'].mean())
df['Weight'].fillna(df['Weight'].mean())
df['BMI'].fillna(df['BMI'].mean())
df['Blood Pressure'].fillna(df['Blood Pressure'].mean())
df['Heart Rate'].fillna(df['Heart Rate'].mean())
df['Temperature'].fillna(df['Temperature'].median())
df['Cholesterol'].fillna(df['Cholesterol'].median())
df['Diabetes'].fillna(df['Diabetes'].mode())
df['Gender'].fillna(df['Gender'].mode())
df['Smoking'].fillna(df['Smoking'].mode())
df['Blood Type'].fillna(df['Blood Type'].mode())
print("Missing values after handling:\n",df.isnull().sum())


#Check for duplicates
duplicate_count=df.duplicated().sum()
print(f"Number of duplicate rows:{duplicate_count}")
if duplicate_count>0:
  df.drop_duplicates(inplace=True)
  print(f"New dataset shape after removing duplicates:{df.shape}")

#Save the cleaned dataset to a new CSV file
df.to_csv('cleaned_medical_students_dataset.csv',index=False)
print("Cleaned dataset saved to 'cleaned_medical_students_dataset.csv'")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#Load dataset of titanic
df=pd.read_csv('/content/drive/MyDrive/Data Set/Copy of Titanic-Dataset.csv')
#Subset and filter for first-class passengers.
subset =df[['Name','Pclass','Sex','Age','Fare','Survived']]
first_class_passengers = subset[subset['Pclass'] == 1]
print("Mean Fare:",first_class_passengers['Fare'].mean())
print("Median Age:",first_class_passengers['Age'].median())
print("Max Fare:",first_class_passengers['Fare'].max())
print("Min Fare:",first_class_passengers['Fare'].min())

#Count null values in age
null_age_count = first_class_passengers['Age'].isnull().sum()
print("Null values in Age:",null_age_count)
#Drop rows with null "Age"
first_class_passengers = first_class_passengers.dropna(subset=['Age'])
print("Rows after dropping null values in Age:",len(first_class_passengers))

#one-hot encoding to embarked.
df =pd.get_dummies(df , columns=['Embarked'], prefix='Embarked',drop_first='False')
print(df.head())

#Calculate survival rates by gender.
survival_rate_by_gender = df.groupby('Sex')['Survived'].mean()
print("Survival rates by gender:")
print(survival_rate_by_gender)
survival_rate_by_gender.plot(kind='bar',color=['blue','green'])
plt.ylabel('Survival Rate')
plt.title('Survival Rates by Gender')
plt.show()

#Loop for the columns like 'Emnarked_C','Embarked_Q','Embarked_S'.
embarked_columns = [col for col in df.columns if 'Embarked' in col]
print("Embarked Columns after One-Hot Encoding:", embarked_columns)
#Calculate survival rates by port of embarkation.
survival_by_embarked = df.groupby(embarked_columns)['Survived'].mean()
#Display the survival rates by port of embarkation.
print("Survival rates by port of embarkation port:")
print(survival_by_embarked)

#Plot the survival rates by embarkation port.
survival_by_embarked.plot(kind='bar',color=['red','blue','green'])
plt.xlabel('Embarkation Port(C,Q,S)')
plt.ylabel('Survival Rate')
plt.title('Survival Rates by Embarkation Port')
plt.show()
```
